```python
data = {
    'Nama' : ['Ali', 'Masya', 'Roni', 'Cika', 'Deny'],
    'Usia' : [22, 21, 18, 25, 30],
    'Asal' : ['Jakarta', 'Bandung', 'Jakarta', 'Bogor', 'Bogor'],
    'Pekerjaan' : ['Swasta', 'Swasta', 'Swasta', 'Karyawan', 'Karyawan']
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#Perusahaan hanya ingin tahun, nama, dan asal
df[['Nama', 'Pekerjaan']]
```

```python
#Perusahaan hanya ingin pekerja dengan usia dibawah 25
df[(df['Usia'] < 25) & (df['Asal'] == 'Jakarta')]
```
