## Pandas with CSV and Visualization

```python
df = pd.read_csv('covid_19_indonesia_time_series_all.csv')
```

```python
df = df[df["Location"].str.contains("Indonesia") == False]
```

```python
df
```

```python
lcoation_total_case = df.groupby(['Location'])['New Cases'].max()
```

```python
lcoation_total_case.plot(kind='bar')
```

```python
total_case = df.groupby(['Location'])['New Cases', 'New Deaths'].mean()
```

```python
total_case.plot(kind='bar')
```
